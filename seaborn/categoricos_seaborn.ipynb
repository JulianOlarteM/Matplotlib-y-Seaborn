```python
import seaborn as sns 
import matplotlib.pyplot as plt

```

```python
tips = sns.load_dataset('tips')
tips.head()
```

Una forma de concentrar la cantedidad de datos para ciertas variables cagtegoricas 

```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

```python
sns.countplot(data=tips,y='day',hue='sex')
plt.show()
```

Otro comando para distribuciones de varibales categoricas.

```python
sns.stripplot(data=tips, x ='day', y= 'total_bill', hue='sex')
```

```python
plt.figure(figsize=(7,7))
sns.stripplot(data=tips, x ='day', y= 'total_bill', hue='sex', dodge = True)
```

Otro comando que permite ver mejor la acumulacion de datos en ciertos puntos. 

```python
plt.figure(figsize=(7,7))
sns.swarmplot(data=tips, x ='day', y= 'total_bill', hue='sex', dodge = True)
```

la distribucion de los datos por percentiles.

```python
plt.figure(figsize=(7,7))
sns.boxplot(data=tips, y= 'total_bill')
plt.show()
```

Para varibales categoricas y segmentada

```python
plt.figure(figsize=(7,7))
sns.boxplot(data=tips, x ='day', y= 'total_bill', hue='sex', dodge = True)
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.boxplot(data=tips, x ='day', y= 'total_bill', hue='sex', dodge = True)
sns.swarmplot(data=tips, x ='day', y= 'total_bill', hue='sex', dodge = True, color = "0" , marker = "<")
plt.show()
```

Otra manera para deterinar donde se concentran los datos es.

```python
plt.figure(figsize=(7,7))
sns.violinplot(data=tips, x ='day', y= 'total_bill', hue='sex', palette='pastel')
plt.show()
```

combinandolos :

```python
plt.figure(figsize=(7,7))
sns.violinplot(data=tips, x ='day', y= 'total_bill', hue='sex', palette='pastel', split =True)
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.violinplot(data=tips, y= 'total_bill')
plt.show()
```

Una manera de utilizar todas la funcioones categoricas con un solo comando es la siguiente 

```python
sns.catplot(data=tips, x ='day', y= 'total_bill', hue= 'sex', dodge = True)
plt.show()
```

```python
sns.catplot(data=tips, x ='day', y= 'total_bill', hue= 'sex',kind="swarm", dodge = "True")
plt.show()
```

```python
sns.catplot(data=tips, x ='day', y= 'total_bill', hue= 'sex',kind="box", dodge = "True")
plt.show()
```

```python
sns.catplot(data=tips, x ='day', y= 'total_bill', hue= 'sex',kind="box", dodge = "True", col ="time")
plt.show()
```

```python
sns.catplot(data=tips, x ='day', y= 'total_bill', hue= 'sex',kind="boxen")
plt.show()
```

```python
sns.catplot(data=tips, x ='day', y= 'total_bill', hue= 'sex',kind="violin")
plt.show()
```

```python
sns.catplot(data=tips, x ='day', y= 'total_bill', hue= 'sex',kind="point")
plt.show()
```

```python
sns.catplot(data=tips, x ='day', y= 'total_bill', hue= 'sex',kind="bar")
plt.show()
```
