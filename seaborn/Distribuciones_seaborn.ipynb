```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.histplot(data=tips, x= 'tip',bins=15) #bins me permite determinar el numero de barras 
plt.show()
```

Distribucion acumulada

```python
sns.histplot(data=tips, x= 'tip',bins=15,cumulative =True)
plt.show()
```

Distribuciones basadas en conteo, probabilidad, densidad y porcentaje.

```python
sns.histplot(data=tips, x= 'tip',bins=15,cumulative = False , hue='sex', stat='count' )
plt.show()

sns.histplot(data=tips, x= 'tip',bins=15,cumulative = False , hue='sex', stat='probability' )
plt.show()

sns.histplot(data=tips, x= 'tip',bins=15,cumulative = False , hue='sex', stat='density' )
plt.show()

sns.histplot(data=tips, x= 'tip',bins=15,cumulative = False , hue='sex', stat='percent' )
plt.show()
```

Como quiero presentar mis datos

```python
sns.histplot(data=tips, x= 'tip',bins=15,cumulative = False , hue='sex', stat='count', multiple='stack' )
plt.show()

sns.histplot(data=tips, x= 'tip',bins=15,cumulative = False , hue='sex', stat='count', multiple='layer' )
plt.show()

sns.histplot(data=tips, x= 'tip',bins=15,cumulative = False , hue='sex', stat='count', multiple='dodge' )
plt.show()
```

Otro comando para graficas de distribucion. kde(Diagrama de densidad)


```python
sns.kdeplot(data=tips, x='tip', hue = 'sex')
```

```python
sns.kdeplot(data=tips, x='tip', hue = 'sex', cumulative = True)

```

Para dibujar el area bajo la curva

```python
sns.kdeplot(data=tips, x='tip', hue = 'sex', cumulative = True, shade=True)
```

```python
sns.kdeplot(data=tips, x='tip', hue = 'sex', cumulative = False, shade=True)
```

Para ajustar los datos.

```python
sns.kdeplot(data=tips, x='tip', hue = 'sex', cumulative = False, shade=True,  bw_adjust=2)
```

Otro grafico: Grfico escalonado de la proporcion que tienen las propinas a lo largo de su frecuencia.

```python
sns.ecdfplot(data=tips, x='tip', hue = 'sex')
```

puedo decirle que no quiero la proporcion sino un conteo. 

```python
sns.ecdfplot(data=tips, x='tip', hue = 'sex', stat ='count')
```

Es importante tener en cuenta que cn displot es posuble realizar todas estas grafivas de manera sencilla

```python
sns.displot(data=tips, x='tip', hue = 'sex', kind='kde')
```

```python
sns.displot(data=tips, x='tip', hue = 'sex', kind='kde', cumulative = True)
```

```python
sns.displot(data=tips, x='tip', hue = 'sex', kind='hist', cumulative = True, multiple = 'stack')
```

```python
sns.displot(data=tips, x='tip', hue = 'sex', kind='hist', cumulative = False, multiple = 'stack')
```
