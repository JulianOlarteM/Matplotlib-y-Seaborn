```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

traemos datos de seaborn , esta tiene datasets precargados para realizar ejemplos de graficacion  

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill',  y= 'tip')
```

```python
sns.displot(data=tips, x='total_bill',  y= 'tip', hue  ='sex' )
```

```python
sns.displot(data=tips, x='total_bill', hue  ='sex' ) # el parametro hue me permite segmentar mis datos 
```

```python
sns.displot(data = tips, x='total_bill', hue='sex', kind ='kde', legend =True, palette = 'dark', alpha = 0.5) #king = 'kde' me permite que la grafica de dispercion sea por densidad.
```
