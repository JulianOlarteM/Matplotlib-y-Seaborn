```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head(3)
```

ESTA FUNCION ME PERMITE TENER DOS GRAFICOS AL MISMO TIEMPO. Un grafico de relacion y uno de dispersion.

```python
sns.jointplot(data = tips, x= 'total_bill', y = 'tip')
plt.show()
```

```python
sns.jointplot(data = tips, x= 'total_bill', y = 'tip', hue = 'sex')
plt.show()
```

Puedo definir tambien el tipo de grafico que quiero

```python
sns.jointplot(data = tips, x= 'total_bill', y = 'tip', hue = 'sex', kind ='hist')
plt.show()
```

puedo definir que la egunda grafica tenga o no un eje definido

```python
sns.jointplot(data = tips, x= 'total_bill', y = 'tip', hue = 'sex', kind ='hist', marginal_ticks=True)
plt.show()
```

Puedo tambien definir las caracteristicas de mi segunda grafica. 

```python
sns.jointplot(data = tips, x= 'total_bill', y = 'tip', hue = 'sex', kind ='hist', marginal_ticks=True
              , marginal_kws = dict(bins =25, fill=False, multiple='dodge', ))
plt.show()
```

Se puede definir el tipo de grafica que queremos. 

```python
sns.jointplot(data = tips, x= 'total_bill', y = 'tip', hue = 'sex', kind ='kde')
plt.show()
```

UNA MANERA DE REALCIONAR TODAS LAS VARIABLES NUMERICAS DE UN DATA SET ES :

```python
sns.pairplot(data=tips)
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='Purples_r')
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='PuOr_r', corner=True)
plt.show()
```
