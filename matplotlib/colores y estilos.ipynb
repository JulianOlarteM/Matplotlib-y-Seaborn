```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
x = np.linspace(0,5,11)
```

Para tener una idea de los estilo predefinidos en matplotlib se tiene:

```python
print(plt.style.available)
```

Aplicamos un estilo definido a la grafica:

```python
plt.style.use('dark_background')
```

Matlab

```python
fig , ax = plt.subplots(figsize = (8,8))

ax.plot(x,x+1,'ro--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'yx-')
ax.plot(x,x+4,'go:')
```

Pyplot

```python
fig , ax = plt.subplots(figsize = (8,8))

ax.plot(x,x+1, color = 'green', alpha = 0.5, linewidth=5 , linestyle = '-' , marker = 'o', markersize= 8 , markerfacecolor = '#CCFFCC' )
ax.plot(x,x+2, color = 'blue', alpha = 0.3 ,linewidth=5,linestyle = '--' , marker = '8', markersize= 10 , markerfacecolor = '#66FF66' )
ax.plot(x,x+3, color = '#66FF66',linestyle = 'dashed' , marker = 'v')
ax.plot(x,x+4, color = '#CCFFCC', linestyle = ':' , marker = 'P')
```
