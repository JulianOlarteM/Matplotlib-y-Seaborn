```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.subplot(1,2,1)      #(num filas,num columnas, index en el que estoy) al crear una fila y dos colunmas se crean 2 index.
plt.plot(x,y,'rs-')
plt.plot(y,x,'b:')

plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)      #(num filas,num columnas, index en el que estoy) al crear una fila y dos colunmas se crean 2 index.
plt.plot(x,y,'rs-')
plt.plot(y,x,'b:')

plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```
