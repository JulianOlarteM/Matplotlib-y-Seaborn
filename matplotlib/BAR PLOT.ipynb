```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
country = ['INDIA','JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000,800,900,1000,300]
```

```python
plt.bar(country,population,width = 0.5 , color = ['red', 'blue' ])
plt.show()
```

```python
plt.bar(country,population,width = 0.5 , color = ['red', 'blue' ])
plt.xticks(np.arange(5), ('India', 'Japon','Mexico', 'Colombia','Alemania'), rotation=45) #funcion para editar las caracteristicas del eje x.
plt.show()
```

```python
plt.barh(country,population)
plt.show()
```
