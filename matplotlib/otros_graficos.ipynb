```python
 import matplotlib.pyplot as plt 
import numpy as np 
```

```python
data= np.random.randint(1,50,100)
```

Hitograma: popular para distribuciones y frecuencias de ocurrencia . 

```python
plt.hist(data, bins =10, histtype ='bar' )
plt.show()
```

```python
plt.hist(data, bins =10, histtype ='step' )
plt.show()
```

Boxplot o grafica de cajas 

```python
data = np.append(data,200) #agrego un dato fuera que es un fliers, que esta muy alejado de los otros datos 
plt.boxplot(data, vert = False , patch_artist = True , notch = True,showfliers = False )
plt.show()
```

Scatter o dispercion

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30*np.random.rand(N))**2
colors = np.random.rand(N)

```

```python
plt.scatter(x,y,area, c= colors, marker='p', alpha= 0.5)
```
